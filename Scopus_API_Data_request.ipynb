```python
import requests
import matplotlib.pyplot as plt

# Your Scopus API Key
API_KEY = 'd628b47043a284c954dfe8dd5ff80a7f'

# Scopus Search API base URL
BASE_URL = 'https://api.elsevier.com/content/search/scopus'

# Fields and queries
fields = {
    "Solar": "digital twin AND solar",
    "Wind": "digital twin AND wind",
    "Hydroenergy": "digital twin AND hydro",
    "Geothermal": "digital twin AND geothermal",
    "Biomass": "digital twin AND biomass"
}

# Headers for Scopus API requests
headers = {
    'Accept': 'application/json',
    'X-ELS-APIKey': API_KEY
}

# Fetch results
results = {}


for field, query in fields.items():
    params = {
        'query': query,
        'count': 1  # We only need total count
    }
    print(f"Querying: {query}")
    response = requests.get(BASE_URL, headers=headers, params=params)
    data = response.json()

    try:
        total = int(data['search-results']['opensearch:totalResults'])
    except Exception as e:
        print(f"Error for {field}: {e}")
        total = 0

    results[field] = total

# Plot results with external legend
labels = list(results.keys())
sizes = list(results.values())
colors = plt.cm.tab20.colors[:len(labels)]  # Set of visually distinct colors

fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(
    sizes, labels=None, autopct='%1.1f%%', startangle=140, colors=colors
)

# Add a separate legend
ax.legend(wedges, labels, title="Energy Fields", loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Digital Twin Research Papers by Renewable Energy Field (Scopus)")
plt.tight_layout()
plt.axis('equal')
plt.show()
print(results)
```
